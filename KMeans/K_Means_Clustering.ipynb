```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_blobs
```

```python
X,y=make_blobs(n_samples=500,centers=4,random_state = 42)
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
X
```

```python
X.shape,y.shape
```

```python
from sklearn.cluster import KMeans
```

```python
km=KMeans(n_clusters=4)
```

```python
km.fit(X)
```

```python
y_pred=km.predict(X)
```

```python
y_pred
```

```python
plt.scatter(X[:,0],X[:,1],c=y_pred)
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],c='red',s=100)
```

#Noise Data

```python
X,y=make_blobs(n_samples=500,centers=4,cluster_std=2,random_state=40)
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
from sklearn.cluster import KMeans
km=KMeans(n_clusters=4)
km.fit(X)
y_pred=km.predict(X)
plt.scatter(X[:,0],X[:,1],c=y_pred)
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],c='red',s=100)
```

```python
km.cluster_centers_
```

```python
error = []

for k in range(1,20):
  km= KMeans(n_clusters=k)
  km.fit(X)
  error.append(km.inertia_)
```

```python
plt.plot(range(1,20),error) #acha wala k kese le uske liye graph
```
