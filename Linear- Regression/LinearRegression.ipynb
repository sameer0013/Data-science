```python
#internal working of sklearn.linear regres
class CustomRl():
  def __init__(self):
    self.m=None
    self.c=None
  def fit(self,x,y):
    p=0
    q=0
    for i in range(x.shape[0]):
      p=p + ( y[i] - y.mean() ) * ( x[i] - x.mean() )
      q= q + ( x[i] - x.mean() ) **2
    self.m=p/q
    self.c=y.mean() - self.m * x.mean()
  def predict(self,X_test):
    return self.m * X_test + self.c


```

#Multivariant LR

```python
import numpy as np
```

```python
from sklearn.preprocessing import MinMaxScaler #normalizing
mms = MinMaxScaler()
x_train = mms.fit_transform(x_train)
```

```python
x_train
```

```python
#gradient descent
class MyLinearRegression():
  def __init__(self,learning_rate,iteration):
    self.learning_rate=learning_rate
    self.iteration=iteration

  def fit(self,X,Y):
    row,col=X.shape
    self.W=np.random.rand(col)
    self.c=np.random.randint(0,10)
    for i in range(self.iteration):
      y_pre=self.predict(X)     #ycap=w(t)x+c

      dW = -2 * (X.T).dot(Y - y_pre) / row
      dc = -2 * (Y - y_pre).sum() / row

      self.W=self.W-self.learning_rate*dW
      self.c=self.c-self.learning_rate*dc

      plt.scatter(X,Y)
      plt.plot(X,y_pre,color='r')
      plt.title("Iteration:"+str(i))
      plt.show()

  def predict(self,X_test):
    return X_test.dot(self.W) + self.c  #y=w(t)x+c
```

#Polynomial Regression

```python
x=np.linspace(-10,15,200)
#y=2x^2+3x+4
y= 2 * x**2 + 3*x + 4 + np.random.randn(200) * 50
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

```python
x=x.reshape(-1,1)
```

```python
lr= LinearRegression()
lr.fit(x,y)
y_predict=lr.predict(x)
plt.scatter(x,y)
plt.scatter(x,y_predict)
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
po=PolynomialFeatures(degree=2)
X=po.fit_transform(x)
X.shape
```
