```python
x= np.arange(6)
y=x%2
```

```python
y
```

```python
plt.scatter(x,y,c=y)
```

```python
X = x.reshape(-1,1)
```

```python
X
```

```python
from sklearn.tree import DecisionTreeClassifier
model =DecisionTreeClassifier()
model.fit(X,y)
```

```python
model.score(X,y)
```

```python
from sklearn import tree
```

```python
plt.figure(figsize=(10,10))
tree.plot_tree(model,filled=True)
plt.show()
```
